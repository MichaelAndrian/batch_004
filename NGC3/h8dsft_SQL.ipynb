# 3. Create Student Info

```python
sql_createStudentInfo ="""
CREATE TABLE studentInfo (
    code_module varchar(45),
    code_presentation varchar(45),
    id_student INT,
    gender varchar(3),
    imd_band varchar(16),
    highest_education varchar(45),
    age_band varchar(16),
    num_of_prev_attempts INT,
    studied_credits INT,
    region varchar(45),
    disability varchar(3),
    final_result varchar(45),
    FOREIGN KEY (code_module) REFERENCES studentRegistration(code_module),
    FOREIGN KEY (code_presentation) REFERENCES studentRegistration(code_presentation),
    FOREIGN KEY (code_presentation) REFERENCES studentRegistration(code_presentation),
    FOREIGN KEY (id_student) REFERENCES studentRegistration(id_student),
    FOREIGN KEY (id_student) REFERENCES studentAssessment(id_student)
);
"""

curr.execute(sql_createStudentInfo)
conn.commit()

```

# 6. Courses

```python
drop_courses = """
    DROP TABLE courses;

"""
curr.execute(drop_courses)
conn.commit()
```

```python
sql_createCourses = """
CREATE TABLE courses (
    code_module varchar(45),
    code_presentation varchar(45),
    module_presentation_length INT,
    
    FOREIGN KEY (code_module) REFERENCES studentRegistration(code_module),
    FOREIGN KEY (code_presentation) REFERENCES studentRegistration(code_presentation),
    
    FOREIGN KEY (code_module) REFERENCES Assessment(code_module),
    FOREIGN KEY (code_presentation) REFERENCES Assessment(code_presentation),
    
    FOREIGN KEY (code_module) REFERENCES studentInfo(code_module),
    FOREIGN KEY (code_presentation) REFERENCES studentInfo(code_presentation),
    
    FOREIGN KEY (code_module) REFERENCES vle(code_module),
    FOREIGN KEY (code_presentation) REFERENCES vle(code_presentation)

);

"""

curr.execute(sql_createCourses)
conn.commit()
```

```python
courses
```

```python
#Add Data
insert_courses ="INSERT INTO courses(code_module, code_presentation, module_presentation_length) VALUES" 

for index, row in courses.iterrows():

    if index == ((courses.shape[0])-1):
        insert_courses += f"""("{row[0]}", "{row[1]}", {row[2]});"""
    else:
        insert_courses += f"""("{row[0]}", "{row[1]}", {row[2]}),"""

curr.execute(insert_courses)
conn.commit()
```

```python
fetch_courses = """
    SELECT * FROM courses;

"""
curr.execute(fetch_courses)
curr.fetchall()
```
