## 1. Create Student Registration Table

```python
sql_createStudentRegistration ="""
CREATE TABLE studentRegistration (
    code_module varchar(45),
    code_presentation varchar(45),
    id_student INT,
    date_registration INT,
    date_unregistration INT
);
"""

curr.execute(sql_createStudentRegistration)
conn.commit()


```

```python
studentRegistration.columns.values.tolist()
```

```python
# Delete Data

delete_studentRegistration = """
    DELETE FROM studentRegistration;

"""
curr.execute(delete_studentRegistration)
conn.commit()
```

```python
#Add Data
insert_studentRegistration ="""
    INSERT INTO studentRegistration(code_module,code_presentation,id_student,date_registration,date_unregistration)
    VALUES ("{code_module}","{code_presentation}",{id_student},{date_registration},{date_unregistration}); """

for index, row in studentRegistration.iterrows():
#     print(insert_studentRegistration.format(code_module=row['code_module'], code_presentation=row['code_presentation'], id_student=row['id_student'], date_registration=row['date_registration'], date_unregistration=row['date_unregistration']))

    curr.execute(insert_studentRegistration.format(code_module=row['code_module'], code_presentation=row['code_presentation'], id_student=row['id_student'], date_registration=row['date_registration'], date_unregistration=row['date_unregistration']))
    conn.commit()
```

```python
query_studentRegistration ="SELECT * FROM studentRegistration WHERE code_module = 'GGG' "
curr.execute(query_studentRegistration)
curr.fetchall()
```

# 2. Create Student Info

```python
sql_createStudentInfo ="""
CREATE TABLE studentInfo (
    code_module varchar(45),
    code_presentation varchar(45),
    id_student INT,
    gender varchar(3),
    imd_band varchar(16),
    highest_education varchar(45),
    age_band varchar(16),
    num_of_prev_attempts INT,
    studied_credits INT,
    region varchar(45),
    disability varchar(3),
    final_result varchar(45),
    FOREIGN KEY (code_module) REFERENCES studentRegistration(code_module),
    FOREIGN KEY (code_presentation) REFERENCES studentRegistration(code_presentation),
    FOREIGN KEY (code_presentation) REFERENCES studentRegistration(code_presentation),
    FOREIGN KEY (id_student) REFERENCES studentRegistration(id_student),
    FOREIGN KEY (id_student) REFERENCES studentAssessment(id_student)
);
"""

curr.execute(sql_createStudentInfo)
conn.commit()

```

```python
#add foreign key
query_studentInfo = "ALTER TABLE studentInfo FOREIGN KEY (id_student) REFERENCES studentAssessment(id_student);"
curr.execute(query_studentInfo)
curr.fetchall()
```

```python
# Delete Data

delete_studentInfo = """
    DELETE FROM studentInfo;

"""
curr.execute(delete_studentInfo)
conn.commit()
```

```python
studentInfo
```

```python
#Add Data
insert_studentInfo ="""
    INSERT INTO studentInfo(code_module,code_presentation,id_student,gender,region, highest_education,imd_band, age_band, num_of_prev_attempts,studied_credits, disability, final_result)
    VALUES ("{code_module}","{code_presentation}",{id_student},"{gender}", "{region}", "{highest_education}","{imd_band}", "{age_band}", {num_of_prev_attempts},{studied_credits}, "{disability}", "{final_result}"); """

for index, row in studentInfo.iterrows():
#     print(insert_studentRegistration.format(code_module=row['code_module'], code_presentation=row['code_presentation'], id_student=row['id_student'], date_registration=row['date_registration'], date_unregistration=row['date_unregistration']))

    curr.execute(insert_studentInfo.format(code_module=row['code_module'], code_presentation=row['code_presentation'], id_student=row['id_student'], gender = row['gender'], region = row['region'], highest_education = row['highest_education'], imd_band = row['imd_band'], age_band = row['age_band'], num_of_prev_attempts =row['num_of_prev_attempts'], studied_credits = row['studied_credits'], disability = row['disability'], final_result = row['final_result']))
    conn.commit()
```

```python
query_studentInfo = " SELECT * FROM studentInfo"
curr.execute(query_studentInfo)
curr.fetchall()
```

# 3. Student Assessment

```python
sql_createStudentAssessment ="""
CREATE TABLE studentAssessment (
    id_student INT,
    id_assessment INT,
    date_submitted INT,
    is_banked TINYINT,
    score FLOAT
);
"""

curr.execute(sql_createStudentAssessment)
conn.commit()
```

```python
# Delete Data

delete_studentAssessment = """
    DELETE FROM studentAssessment;

"""
curr.execute(delete_studentAssessment)
conn.commit()
```

```python
#Add Data
insert_studentAssessment ="""
    INSERT INTO studentAssessment(id_assessment,id_student, date_submitted, is_banked, score)
    VALUES ({id_assessment},{id_student},{date_submitted},{is_banked}, {score}); """



for index, row in studentAssessment.iterrows():
#     print(insert_studentRegistration.format(code_module=row['code_module'], code_presentation=row['code_presentation'], id_student=row['id_student'], date_registration=row['date_registration'], date_unregistration=row['date_unregistration']))

    curr.execute(insert_studentAssessment.format(id_assessment=row['id_assessment'], id_student=row['id_student'], date_submitted =row['date_submitted'], is_banked = row['is_banked'], score = row['score']))
    conn.commit()
```
