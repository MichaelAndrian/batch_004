# 3. Prepare Data

### Turn it into Greyscale so we have one dimension

```python
image_grey = ImageOps.grayscale(image)
image_grey
```

```python
image_grey_data = np.array(image_grey)
```

```python
image_grey_data.shape
```

# 4. Calculate Gradient

```python
dx, dy = np.gradient(image_grey_data)
```

# 5. Calculate Gradient Magnitude

The formula are given sqrt(dx^2,dy^2)

```python
magnitude = np.sqrt(dx**2,dy**2)
```

```python
magnitude
```

# 6. Turn Magnitude With Threshold ~ 30

```python
magnitude_filtered = np.where(((magnitude > 25) & (magnitude < 35)) , 255, 0)
```

we use the range from 25 to 35

```python
magnitude_filtered
```

```python
plt.imshow(magnitude_filtered)
```

You can see that the edge is detected
